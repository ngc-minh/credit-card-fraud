viết cho tôi đoạn mã để tìm tham số tối ưu của mô hình Logistic Regression sao cho có giá trị f1 score tốt nhất và sử dụng cross calidation và thuật toán tối ưu bayesian optimization

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, f1_score
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

# Định nghĩa mô hình Logistic Regression
model = LogisticRegression(solver='liblinear')

# Định nghĩa không gian tìm kiếm tham số
search_spaces = {
    'C': Real(1e-6, 1e+6, prior='log-uniform'),
    'penalty': Categorical(['l1', 'l2']),
}

# Định nghĩa cross-validation và bộ tối ưu hóa Bayesian
opt = BayesSearchCV(
    model,
    search_spaces,
    n_iter=50,
    scoring=make_scorer(f1_score),
    cv=5,
    n_jobs=-1,
    verbose=0,
    random_state=42
)

# Thực hiện tìm kiếm tham số
opt.fit(X, y)

# In ra kết quả tốt nhất
print("Best parameters:", opt.best_params_)
print("Best F1 Score:", opt.best_score_)
```
